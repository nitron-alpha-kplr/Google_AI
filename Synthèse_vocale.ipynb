```
from gtts import gTTS
from IPython.display import Audio
```

```
speech = gTTS('hello there, the voice synthesis is perfectly working ')
speech.save('my_line.wav')
sound_file = 'my_line.wav'
Audio(sound_file, autoplay=True)
```

```
def read_aloud(phrase,nom_fichier):
    speech = gTTS(phrase)
    speech.save(nom_fichier)
    return(Audio(nom_fichier, autoplay=True))
```
